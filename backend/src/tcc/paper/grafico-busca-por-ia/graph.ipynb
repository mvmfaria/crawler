```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ler o arquivo .csv
data = pd.read_csv('./ai-search.csv')


# Extrair as colunas desejadas
x = data['month']
y = data['popularity']

# Gerar o gráfico de linha
plt.figure(figsize=(14, 4))  # Aumenta o tamanho da imagem
plt.plot(x, y)
plt.xlabel('Data')
plt.ylabel('Popularidade')
plt.title('Busca por IA no Google nos últimos 7 anos.')

# Definir os locais e rótulos dos marcadores do eixo x
xticks_locs = np.arange(0, len(x), 12)  # Atualiza o passo para 6
xticks_labels = x[::12]  # Atualiza o passo para 6
plt.xticks(xticks_locs, xticks_labels)  # Define os locais e rótulos dos marcadores do eixo x

# Exibir o gráfico
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Aumenta o tamanho da imagem
plt.scatter(x, y)
plt.xlabel('Mês')
plt.ylabel('Popularidade')
plt.title('Busca por IA no Google ao longo dos anos.')
plt.xticks(rotation='vertical')  # Rotaciona os rótulos do eixo x
plt.show()
```
